```python
def new_limit(ma):
    if ma<5e-11:
        return 3e-12*(1+(ma/5e-11)**4)
    else:
        return 3e-12*(1+(ma/5e-11)**4)
    
    
def SN_limit(ma):
    if ma<5e-11:
        return 6e-12*(1+(ma/5e-10)**4)
    else:
        return 6e-12*(1+(ma/5e-10)**4)
```
